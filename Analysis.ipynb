```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plt.style.use('seaborn-v0_8-whitegrid')  # or 'ggplot', 'fivethirtyeight', etc.
plt.style.use('ggplot')  # or 'ggplot', 'fivethirtyeight', etc.
```

```python
# Load later
score_mats = np.load("score_mats.npy", allow_pickle=True)
```

```python
# event names
event_names = ['Keqiao\n2024','SLC\n2024','Innsbruck\n2024',
               'Prague\n2024','Seoul\n2024','Keqiao\n2025',
               'Curitiba\n2025','SLC\n2025']
```

```python
# plot score deviation by athlete
fig,ax = plt.subplots()
Colors = ['red','blue']
labels = ['Men','Women']
avg_v = np.empty((len(score_mats),2))


# plot standard deviation of scores for individual boulders
for i_event, score_mat_event in enumerate(score_mats):
    for i_gender,score_mat in enumerate(score_mat_event):
        athlete_score = np.sum(score_mat[:4,:], axis=1)
        avg_v[i_event,i_gender] = np.std(athlete_score)
        # ax.scatter(np.full_like(athlete_score,i_event), athlete_score, c=Colors[i_gender],alpha=0.5)

# plot average
for i_gender in range(2):
    ax.plot(np.arange(len(score_mats)), avg_v[:,i_gender], c=Colors[i_gender], label =labels[i_gender])
ax.set(xticklabels=event_names, 
       xticks=np.arange(len(event_names)),
       xlabel='Event', 
       ylabel='Standard Deviation of Scores',
       title='Score separation for each boulder')
ax.legend()
plt.savefig('score_separation.png', dpi=300, bbox_inches='tight')
```

```python
fig, ax = plt.subplots()
colors = ['red', 'blue']
labels = ['Men', 'Women']  # Or whatever gender/order applies

# Loop through events and genders
for i_event, score_mat_event in enumerate(score_mats):
    for i_gender, score_mat in enumerate(score_mat_event):
        v_score = np.std(score_mat, axis=0)
        x = np.full_like(v_score, i_event)
        ax.scatter(x, v_score, color=colors[i_gender], label=labels[i_gender] if i_event == 0 else "")

# Add axis labels and title
ax.set_xlabel("Event Index")
ax.set_ylabel("Score Std Dev")
ax.set_title("Score Variability by Event and Gender")

# Handle overlapping labels
ax.legend(loc="best")
ax.grid(True)
plt.tight_layout()
plt.show()
```
